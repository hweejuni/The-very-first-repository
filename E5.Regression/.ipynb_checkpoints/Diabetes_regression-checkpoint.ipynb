```python
df_X, df_y = diabetes.data[:,:10], diabetes.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
W = np.random.rand(len(X_train[0,:])) # set random weights
b = np.random.rand()
```

```python
def model(X, W, b):
    predictions = 0
    for i in range(len(X[0,:])):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions # predictions = b + x_1*w_1 + w_2*x_2 + ... + w_n*x_n 
```

```python
def MSE(a, b):
    mse = ((a - b) ** 2).mean() 
    return mse
```

```python
def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

```python
def gradient(X, W, b, y):
    N = len(X)
    y_pred = model(X, W, b)
    dW = 1/N * 2 * X.T.dot(y_pred - y)
    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
dW, db = gradient(X_train, W, b, y_train)
print("dW:", dW)
print("db:", db)
```

```python
LEARNING_RATE = 0.05
```

```python
losses = []

for i in range(1, 15001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 1000 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```
