```python
import pandas as pd
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
bike = pd.read_csv('~/aiffel/bike_regression/data/bike_train.csv')
print(bike.shape)
pd.DataFrame(bike)
```

datetime - hourly date + timestamp  
season -  1 = spring, 2 = summer, 3 = fall, 4 = winter   
holiday - whether the day is considered a holiday   
workingday - whether the day is neither a weekend nor holiday   
weather - 1: Clear, Few clouds, Partly cloudy, Partly cloudy   
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist   
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds   
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog    
temp - temperature in Celsius   
atemp - "feels like" temperature in Celsius   
humidity - relative humidity   
windspeed - wind speed   
casual - number of non-registered user rentals initiated   
registered - number of registered user rentals initiated   
count - number of total rentals   

```python
bike['year'] = pd.DatetimeIndex(bike['datetime']).year
bike['month'] = pd.DatetimeIndex(bike['datetime']).month
bike['day'] = pd.DatetimeIndex(bike['datetime']).day
bike['hour'] = pd.DatetimeIndex(bike['datetime']).hour
bike['minute'] = pd.DatetimeIndex(bike['datetime']).minute
bike['second'] = pd.DatetimeIndex(bike['datetime']).second

pd.DataFrame(bike)
```

```python
df_X = bike[['season','holiday','workingday','weather','temp','atemp','humidity','windspeed',
                 'casual','registered','year','month','day','hour','minute','second']]
df_y = bike['count']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(predictions)
```

```python
def mse(a, b):
    y = ((a - b)**2).mean()
    return y

def rmse(a ,b):
    y = mse(a, b)**0.5
    return y
```

```python
losses = y_test - predictions
mse_e = mse(predictions, y_test)
rmse_e = rmse(predictions, y_test)

print(losses, mse_e, rmse_e)
```

```python
bike['Error'] = losses
pd.DataFrame(bike)
```
